#### Understanding the Dataset:

Before proceeding further, it is good to know more about the Iris Dataset.

#### Data Set Information:

* This is perhaps the best known database to be found in the pattern recognition literature. Fisher's paper is a classic in the field and is referenced frequently to this day. (See Duda & Hart, for example.) The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant. One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

* Predicted attribute: class of iris plant.

* This is an exceedingly simple domain.


#### Attribute Information:

1. sepal length in cm
2. sepal width in cm
3. petal length in cm
4. petal width in cm
5. class:
 * Iris Setosa
 * Iris Versicolour
 * Iris Virginica
 
 Reference: https://archive.ics.uci.edu/ml/datasets/iris

### 1. Count Plot

```python
#Count of Class Labels - CountPlot

plt.figure(figsize=(7,7))
sns.set_style("whitegrid")
sns.countplot(x='Classification', data=iris_data, palette='autumn')

plt.title("Distribution of Class Labels", fontsize=15)
plt.xlabel("Classification of Species", fontsize=12)
plt.ylabel("Value Count", fontsize=12)
plt.show()
```

#### Observation:
1. There are three class labels namely Iris-setosa, Iris-versicolor, Iris-virginica.
2. The dataset is balanced with 50 datapoints to each classs. (33% each)

### 2. Heatmap - Correlation

```python
iris_data.corr()
```

```python
#Heatmap for Correlation of the Features 
plt.figure(figsize=(8,6))
sns.heatmap(iris_data.corr(), annot=True)
plt.show()
```

#### Observation:

From the Correlation Table and Heatmap, we see that:

* Sepal Length and Sepal Width are weakly correlated [~ -0.109]
* Sepal Length and Petal Length are highly correlated [~ 0.871]
* Sepal Length and Petal Width are highly correlated [~ 0.817]
* Sepal Length and Petal Width are weakly correlated [~ -0.420] and so on.
* The highest correlation is between Petal Length and Petal Width [~ 0.963]

### 3. Scatter Plots

```python
#SW vs SL
plt.style.use('classic')
plt.figure(figsize=(6,4))
plt.scatter(iris_data['SW'],iris_data['SL'],s=10,c='green',marker='o')
plt.title("Scatter plot between Sepal Width and Sepal Length",fontsize=12)
plt.xlabel("Sepal Width (SW)")
plt.ylabel("Sepal Length (SL)")
plt.grid()
plt.show()
```

* It is difficult to make out between different class labels.
* Hence, we distinguish datapoints belonging to different classes with different color.

```python
#1 - SW vs SL
sns.set_style("darkgrid");
sns.FacetGrid(iris_data, hue="Classification", height=6)\
   .map(plt.scatter, "SW", "SL") \
   .add_legend(fontsize=12)
plt.title("Scatter plot between Sepal Width and Sepal Length", fontsize =15)
plt.xlabel("Sepal Width (SW)")
plt.ylabel("Sepal Length (SL)")
plt.show()
```

```python
#3 - SW vs PW
sns.set_style("darkgrid");
sns.FacetGrid(iris_data, hue="Classification", height=6)\
   .map(plt.scatter, "SW", "PW") \
   .add_legend(fontsize=12)
plt.title("Scatter plot between Sepal Width and Petal Width", fontsize =15)
plt.xlabel("Sepal Width (SW)")
plt.ylabel("Petal Width (PL)")
plt.show()
```

```python
#4 - SL vs PL
sns.set_style("darkgrid");
sns.FacetGrid(iris_data, hue="Classification", height=6)\
   .map(plt.scatter, "SL", "PL") \
   .add_legend(fontsize=12)
plt.title("Scatter plot between Sepal Length and Petal Length", fontsize =15)
plt.xlabel("Sepal Length (SL)")
plt.ylabel("Petal Length (PL)")
plt.show()
```

```python
#5 - SL vs PW
sns.set_style("darkgrid");
sns.FacetGrid(iris_data, hue="Classification", height=6)\
   .map(plt.scatter, "SL", "PL") \
   .add_legend(fontsize=12)
plt.title("Scatter plot between Sepal Length and Petal Length", fontsize =15)
plt.xlabel("Sepal Length (SL)")
plt.ylabel("Petal Width (PW)")
plt.show()
```

### 4. Pair Plots 

```python
sns.set_style("darkgrid");
sns.pairplot(iris_data, hue="Classification", height=4);
plt.show()
```

#### Observations:
1. We can see that the Plot between PL and PW produces the maximum separation between the points.
2. Hence we could say that these two features are the most useful for the Classification.
3. The datapoints belonging to the Iris Setosa species are linearly separable, the datapoints belonging to Versicolor and Virginica have considerable overlap.

```python
#BoxPlot of PL and Species

sns.boxplot(x='Classification',y='PL', data=iris_data)
plt.show()
```

#### Observations:
1. PL of Iris_Setosa and Versicolor has outliers.
2. The Median of PL lies at around 1.5 cm for Setosa, 4.2 cm for Versicolor and 5.5 cm for Virginica.
3. IQR of Setosa lies between 0.4 to 0.6, Versicolor between 4.0 to 4.5 and Virginica between 5.1 to 5.9.

```python
#BoxPlot of PW and Species

sns.boxplot(x='Classification',y='PW', data=iris_data)
plt.show()
```

#### Observations:
